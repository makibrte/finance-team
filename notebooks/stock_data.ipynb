```python
stocks_df = stocks_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y', 'minute_x', 'minute_y'], axis=1)
```

```python
x = 0
y = 0
stocks_df_temp = stocks_df.copy(deep=True)
cols = []
for n, col in enumerate(stocks_df_temp.columns):
    if col.split('_')[-1] == 'x':
        cols.append('{}_{}'.format(col, x))
        if n % 5 == 0:
            x += 1
        
    elif col.split('_')[-1] == 'y':
        cols.append('{}_{}'.format(col, y))
        if n % 5 == 0:
            y += 1
    else:
        cols.append(col)
    
```

```python
stocks_df_temp.columns = cols
stocks_df_temp
```

```python
stocks_df_temp['datetime'] = pd.to_datetime(stocks_df_temp['datetime'])
stocks_df_temp['date'] = pd.to_datetime(stocks_df_temp['datetime'].dt.date)
stocks_df_temp['hour'] = stocks_df_temp['datetime'].dt.hour + stocks_df_temp['datetime'].dt.minute / 60
stocks_df_temp
```
